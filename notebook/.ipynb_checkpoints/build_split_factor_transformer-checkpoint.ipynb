```python
import pandas as pd
import numpy as np
import os
import joblib
import seaborn as sns
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.neighbors import NearestNeighbors
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
import plotly.express as px
import warnings;warnings.simplefilter('ignore')
root = '../data/phase_2/'
os.listdir(root)
```

```python
col_names = joblib.load('../data/phase_2/col_names.pkl')
col_names.keys()
```

```python
def get_col(df,name):
    return df.columns[df.columns.str.contains(name)].tolist()
```

```python
def load_data(path):
    df = pd.read_excel(path)
    df = df.dropna(axis=0,how='all').dropna(axis=1,how='all').reset_index(drop=True)
    df.columns = [i for i in range(df.shape[1])]
    df.loc[:,0] = df.loc[:,0].fillna(method='ffill')
    df = df.T
    cols = []
    for i in range(df.shape[1]):
        name = ''
        for j in range(3):
            name += str(df.loc[j,i]).strip()+'_'
        cols.append(name)
    cols = [i.replace('nan','').strip('_') for i in cols]
    cols = [i.strip() for i in cols]
    df.columns = cols
    df = df.iloc[4:,1:].reset_index(drop=True)
    df = df.drop(df.columns[(df.isnull().sum() > 0).values].tolist(),axis=1)
    print(df.shape)
    return df.astype('float')
```

# load data

```python
df = load_data(root+'Aroma-1 Unit 000 Split_Factor Calculation.xlsx')
```

```python
df.head()
```

# define columns

```python
Fna_col = [get_col(df,'Naphtha Properties')[1]]
Xna_col = get_col(df,'Naphtha Properties')[5:5+54]
#===============================================
Sle_col = get_col(df,'Light End Split Factor')
Shc_col = get_col(df,'Heart Cut Split Factor')
She_col = get_col(df,'Heavy End Split Factor')
#===============================================
Fle_col = [get_col(df,'Light End Product Properties')[1]]
Xle_col = get_col(df,'Light End Product Properties')[3:3+54]
#===============================================
Fhc_col = [get_col(df,'Heart Cut Product Properties')[1]]
Xhc_col = get_col(df,'Heart Cut Product Properties')[4:4+54]
#===============================================
Fhe_col = [get_col(df,'Heavy End Product Properties')[1]]
Xhe_col = get_col(df,'Heavy End Product Properties')[3:3+54]
```

```python
col_names['sle'] = Sle_col
col_names['shc'] = Shc_col
col_names['she'] = She_col
```

# rename columns

```python
z = {**dict(zip(Xna_col,col_names['xna'])), 
     **dict(zip(Fna_col,col_names['fna'])),
     **dict(zip(Xle_col,col_names['xle'])),
     **dict(zip(Fle_col,col_names['fle'])),
     **dict(zip(Xhc_col,col_names['xhc'])),
     **dict(zip(Fhc_col,col_names['fhc'])),
     **dict(zip(Xhe_col,col_names['xhe'])),
     **dict(zip(Fhe_col,col_names['fhe']))}
df = df.rename(columns=z)
df.head()
```

# build transformer

```python
class split_factor_transformer(object):
    def __init__(self):
        self.Xna_col = col_names['xna']
        self.Fna_col = col_names['fna']
        #=====================
        self.Sle_col = col_names['sle']
        self.Shc_col = col_names['shc']
        self.She_col = col_names['she']
        #=====================
        self.Xle_col = col_names['xle']
        self.Xhc_col = col_names['xhc']
        self.Xhe_col = col_names['xhe']
        #=====================
        self.Fle_col = col_names['fle']
        self.Fhc_col = col_names['fhc']
        self.Fhe_col = col_names['fhe']
    
    def __call__(self,df):
        sle = self.calculate_sp(df,self.Xle_col,self.Fle_col,self.Sle_col)
        shc = self.calculate_sp(df,self.Xhc_col,self.Fhc_col,self.Shc_col)
        she = self.calculate_sp(df,self.Xhe_col,self.Fhe_col,self.She_col)
        #=================
        sle = sle.fillna(1)
        sle.iloc[:,-1] = 0
        #=================
        shc = shc.fillna(0)
        #=================
        she = she.fillna(0)
        she.iloc[:,-1] = 1
        return sle,shc,she
    
    def calculate_sp(self,df,x_col,f_col,s_col):
        Xna = df[self.Xna_col].values
        Fna = df[self.Fna_col].values.reshape(-1,1)
        Xw = df[x_col].values
        Fw = df[f_col].values.reshape(-1,1)
        return pd.DataFrame((Xw*Fw)/(Xna*Fna),columns=s_col)
```

```python
sp_tr = split_factor_transformer()
```

# make prediction

```python
sle,shc,she = sp_tr(df)
```

```python
print(round((sle - df[Sle_col]).sum().sum(),12))
sle.head(1).append(df[Sle_col].head(1))
```

```python
print(round((shc - df[Shc_col]).sum().sum(),12))
shc.head(1).append(df[Shc_col].head(1))
```

```python
print(round((she - df[She_col]).sum().sum(),12))
she.head(1).append(df[She_col].head(1))
```
