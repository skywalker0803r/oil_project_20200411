```python
from API.utils import *
import pandas as pd
import numpy as np
import os
import seaborn as sns
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.neighbors import NearestNeighbors
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
import plotly.express as px
import joblib
import warnings;warnings.simplefilter('ignore')
root = './data/phase_2/'
os.listdir(root)
```

```python
def get_col(df,name):
    return df.columns[df.columns.str.contains(name)].tolist()
```

```python
os.listdir('model')
```

```python
col_names = joblib.load('./data/phase_2/col_names.pkl')
col_names.keys()
```

# Load data

```python
train = pd.read_csv(root+'train.csv',index_col=0)
train = train.dropna(axis=1,thresh=14000)
train = train.dropna(axis=0)
train = train.astype('float')
print(train.shape)
train.head()
```

```python
case_col = set(get_col(train,'Case'))&set(get_col(train,'Input'))
case_col = sorted(list(case_col))
col_names['case'] = case_col
joblib.dump(col_names,'./data/phase_2/col_names.pkl')
case_col
```

```python
train = train[case_col+col_names['all']]
train.shape
```

# sle

```python
a = train[col_names['xle']].values
b = train[col_names['fle']].values
c = train[col_names['xna']].values
d = train[col_names['fna']].values
sle = (a*b)/(c*d)
sle = pd.DataFrame(sle,columns=col_names['sle'])
sle = sle.fillna(1)
sle.iloc[:,-1] = 0
sle.head()
```

# shc

```python
a = train[col_names['xhc']].values
b = train[col_names['fhc']].values
c = train[col_names['xna']].values
d = train[col_names['fna']].values
shc = (a*b)/(c*d)
shc = pd.DataFrame(shc,columns=col_names['shc'])
shc = shc.fillna(0)
shc.head()
```

# she

```python
a = train[col_names['xhe']].values
b = train[col_names['fhe']].values
c = train[col_names['xna']].values
d = train[col_names['fna']].values
she = (a*b)/(c*d)
she = pd.DataFrame(she,columns=col_names['she'])
she = she.fillna(0)
she.iloc[:,-1] = 1
she.head()
```

# add sp factor

```python
print(train.shape)
print(sle.shape,shc.shape,she.shape)
```

```python
cols = train.columns.tolist()
train = pd.DataFrame(np.hstack((train.values,sle.values,shc.values,she.values)))
train.columns = cols + col_names['sle'] + col_names['shc'] + col_names['she']
print(train.shape)
train.head()
```
